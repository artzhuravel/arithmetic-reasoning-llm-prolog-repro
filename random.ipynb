```python
import math

def euler(f, y0, k, n):
    """
    This function takes the right-hand side of the differential eequation dy/dt = f(t,y),
    the initial condition y0 = y(t0), the step size k, and the number of steps n.
    """
    t = 0.0
    y = y0
    for i in range(n):
        # Add your implementation here
        y = y + k * f(t, y)
        t = t + k
    return y

def f(t, y):
    return y - math.sin(t)

y0 = 0.0
k = 1e-2
n = 100   # so t = n*k = 1
print(euler(f, y0, k, n))
```
